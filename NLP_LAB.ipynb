EXP 01 -Write program demonstrates how to use regular expressions in Python to match and search for
patterns in text.

```python
import re

text = "My phone number is 9876543210 and my email is student@gmail.com"

match_result = re.match(r"My", text)
if match_result:
    print("Match Found:", match_result.group())
else:
    print("No Match Found")

search_result = re.search(r"\d{10}", text)
if search_result:
    print("Phone Number Found:", search_result.group())
else:
    print("Phone Number Not Found")

email = re.search(r"\S+@\S+\.\S+", text)
if email:
    print("Email Found:", email.group())
else:
    print("Email Not Found")
```

EXP 02 -Implement a basic finite state automaton that recognizes a specific language or pattern. In this
example, we'll create a simple automaton to match strings ending with 'ab' using python.

```python
def fsa_ends_with_ab(string):
    state = 0

    for ch in string:
        if state == 0:
            if ch == 'a':
                state = 1
            else:
                state = 0

        elif state == 1:
            if ch == 'a':
                state = 1
            elif ch == 'b':
                state = 2
            else:
                state = 0

        elif state == 2:
            if ch == 'a':
                state = 1
            elif ch == 'b':
                state = 0
            else:
                state = 0

    if state == 2:
        print("Accepted")
    else:
        print("Rejected")

# Input
s = input("Enter a string: ")
fsa_ends_with_ab(s)
```

EXP 04 -Implement a finite-state machine for morphological parsing. In this example, we'll create a
simple machine to generate plural forms of English nouns using python.

```python
def pluralize(noun):
    if noun.endswith(('s', 'x', 'z', 'ch', 'sh')):
        return noun + "es"
    elif noun.endswith("y") and noun[-2].lower() not in "aeiou":
        return noun[:-1] + "ies"
    elif noun.endswith("f"):
        return noun[:-1] + "ves"
    elif noun.endswith("fe"):
        return noun[:-2] + "ves"
    else:
        return noun + "s"

word = input("Enter a singular noun: ")
print("Plural form:", pluralize(word))
```

EXP 06 -Implement a basic N-gram model for text generation. For example, generate text using a bigram
model using python.

```python
import random

text = "the cat sat on the mat the cat ate the fish"

words = text.split()
bigrams = {}
for i in range(len(words) - 1):
    key = words[i]
    value = words[i + 1]
    bigrams.setdefault(key, []).append(value)

current = random.choice(list(bigrams.keys()))
generated = [current]

for _ in range(10):
    if current in bigrams:
        current = random.choice(bigrams[current])
        generated.append(current)
    else:
        break

print("Generated Text:")
print(" ".join(generated))
```

EXP 07 -Write program using the NLTK library to perform part-of-speech tagging on a text.

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

text = input("Enter a sentence: ")

words = word_tokenize(text)

pos_tags = nltk.pos_tag(words)

print("\nPart-of-Speech Tags:")
for word, tag in pos_tags:
    print(word, "->", tag)
```

EXP 08 -Implement a simple stochastic part-of-speech tagging algorithm using a basic probabilistic
model to assign POS tags using python.

```python
import random

pos_dict = {
    "the": [("DT", 1.0)],
    "a": [("DT", 1.0)],
    "boy": [("NN", 0.9), ("VB", 0.1)],
    "girl": [("NN", 1.0)],
    "plays": [("VBZ", 0.8), ("NNS", 0.2)],
    "play": [("VB", 0.7), ("NN", 0.3)],
    "cricket": [("NN", 1.0)],
    "runs": [("VBZ", 1.0)],
    "fast": [("RB", 0.7), ("JJ", 0.3)]
}

sentence = input("Enter a sentence: ").lower().split()

print("\nPOS Tags:")
for word in sentence:
    if word in pos_dict:
        tags = [tag for tag, prob in pos_dict[word]]
        probs = [prob for tag, prob in pos_dict[word]]
        tag = random.choices(tags, weights=probs)[0]
    else:
        tag = "NN"  # Default tag
    print(word, "->", tag)
```

EXP 09 -Implement a rule-based part-of-speech tagging system using regular expressions using python.

```python
import re

def pos_tag(word):
    if re.fullmatch(r"(a|an|the)", word):
        return "DT"      # Determiner
    elif re.fullmatch(r"(is|am|are|was|were)", word):
        return "VB"      # Verb
    elif re.fullmatch(r".*ing", word):
        return "VBG"     # Verb (Gerund)
    elif re.fullmatch(r".*ed", word):
        return "VBD"     # Verb (Past Tense)
    elif re.fullmatch(r".*ly", word):
        return "RB"      # Adverb
    elif re.fullmatch(r".*ous|.*ful|.*able|.*ive", word):
        return "JJ"      # Adjective
    elif re.fullmatch(r".*s", word):
        return "NNS"     # Plural Noun
    else:
        return "NN"      # Noun (Default)

sentence = input("Enter a sentence: ").lower()
words = sentence.split()

print("\nPart-of-Speech Tags:")
for word in words:
    print(word, "->", pos_tag(word))
```

EXP 10 -Implement transformation-based tagging using a set of transformation rules, apply a simple rule
to tag words using python.

```python
# Initial dictionary with default POS tags
pos_dict = {
    "the": "DT",
    "a": "DT",
    "an": "DT",
    "boy": "NN",
    "girl": "NN",
    "dog": "NN",
    "cat": "NN",
    "is": "VB",
    "are": "VB",
    "was": "VB",
    "were": "VB"
}

# Input sentence
sentence = input("Enter a sentence: ").lower().split()

print("\nTransformation-Based POS Tagging:")

for word in sentence:
    # Initial tag from dictionary
    tag = pos_dict.get(word, "NN")

    # Transformation rules
    if word.endswith("ing"):
        tag = "VBG"
    elif word.endswith("ed"):
        tag = "VBD"
    elif word.endswith("ly"):
        tag = "RB"
    elif word.endswith("s") and tag == "NN":
        tag = "NNS"

    print(word, "->", tag)
```
